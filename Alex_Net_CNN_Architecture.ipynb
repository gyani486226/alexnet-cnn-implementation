# Alex CNN Architecture


### Step1: Import the required library

```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import layers, models
```

### Step2: Build AlexNet the CNN Architecture

```python

def create_alexnet(input_shape = (227, 227, 3), num_classes = 1000):
    model_alexnet = models.Sequential()


# Layer 1: Conv + MaxPooling
    model_alexnet.add(layers.Conv2D(96, (11,11), strides=(4,4), activation='relu', input_shape=input_shape)) #extracts features.
    model_alexnet.add(layers.MaxPooling2D(pool_size=(3,3), strides=(2,2))) # reduces dimensionality and retains the strongest features.

# Layer 2: Conv + MaxPooling
    model_alexnet.add(layers.Conv2D(256, (5,5), activation='relu', padding='same'))
    model_alexnet.add(layers.MaxPooling2D(pool_size=(3,3), strides=(2,2)))

# Layer 3: Conv
    model_alexnet.add(layers.Conv2D(384, (3,3), activation='relu', padding='same'))

# Layer 4: Conv
    model_alexnet.add(layers.Conv2D(384, (3,3), activation='relu', padding='same'))

# Layer 5: Conv + MaxPooling
    model_alexnet.add(layers.Conv2D(256, (3,3), activation='relu', padding='same'))
    model_alexnet.add(layers.MaxPooling2D(pool_size=(3,3), strides=(2,2)))

# Flatten before Fully Connected layers
    model_alexnet.add(layers.Flatten())
    model_alexnet.add(layers.Dense(4096, activation='relu'))
    # to avoid the overfitting part we use the dropout
    model_alexnet.add(layers.Dropout(0.5))
    model_alexnet.add(layers.Dense(4096, activation = 'relu'))
    # to avoid the overfitting part we use the dropout
    model_alexnet.add(layers.Dropout(0.5))
    model_alexnet.add(layers.Dense(num_classes, activation='softmax'))


    return model_alexnet
```

### Step3: Compile the model

```python
## function calling
model_alexnet_func = create_alexnet()
model_alexnet_func.compile(optimizer = 'adam', loss = "categorical_crossentropy", metrics = ["accuracy"])
model_alexnet_func.summary()
```

### Step4: Save the Model

```python
model_alexnet_func.save("alex_net_mnist_cnn_model.h5")
```
